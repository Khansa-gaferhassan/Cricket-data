```python
import pandas as pd 
import json
```

```python
data =pd.read_csv('match_results.csv')
df =pd.DataFrame(data)
df.head()
```

```python
df.shape
```

```python
match_ids_dict ={}
for index ,row in df.iterrows():
    key1 =row['team1'] + ' Vs ' +row['team2']
    key2 =row['team2'] + ' Vs ' + row['team1']
    match_ids_dict[key1] =row['match_id']
    match_ids_dict[key2] =row['match_id']
    match_ids_dict
```

```python
match_ids_dict
```

**batting summary **

```python
with open('batting_summary.json')as file:
    data =json.load(file)
    all_recordes =[]

    for rec in data:
        all_recordes.extend(rec['battingSummary'])

batting_df =pd.DataFrame(all_recordes)
batting_df.tail()
```

```python
batting_df['out/not_out'] = batting_df.dismissal.apply(lambda x: 'out' if len(x) >0 else 'not_out')
batting_df.tail()
```

batting_df.drop(columns=['dismissal'],inplace=True)
batting_df.sample()

```python
batting_df.head(10)
```

```python
batting_df['batsmanName'] =batting_df['batsmanName'].apply(lambda x:x.replace('†',''))
```

```python
batting_df.head(10)
```

```python
batting_df['match_id'] =batting_df['match'].map(match_ids_dict)
batting_df.head(10)
```

batting_df.to_csv('try.csv',index=False)
